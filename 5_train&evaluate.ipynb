# Training and Evaluating of the Models

```python
import pandas as pd
X_train = pd.read_csv('data/X_train.csv')
X_test = pd.read_csv('data/X_test.csv')
y_train = pd.read_csv('data/y_train.csv')
y_test = pd.read_csv('data/y_test.csv')
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_gridLR = [{
    'max_iter': [100, 200, 300],
    'solver': ["saga", "lbfgs", "newton-cg"],
    'multi_class' : ["ovr"]
}]
lr_clf = LogisticRegression(random_state=1)
grid_searchLR = GridSearchCV(lr_clf, param_gridLR)
grid_searchLR.fit(X_train, y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

final_modelLR = grid_searchLR.best_estimator_
print("Estimator of LR: "+str(final_modelLR))
y_pred_train = final_modelLR.predict(X_train)
print("Accuracy of LR on training: "+str(accuracy_score(y_train, y_pred_train)))
y_pred = final_modelLR.predict(X_test)
print("Accuracy of LR on testing: "+str(accuracy_score(y_test, y_pred)))
print("Precision of LR: "+str(precision_score(y_test, y_pred,average=None)))
print("F1 of LR: "+str(f1_score(y_test, y_pred,average=None)))
print("Recall of LR: "+str(recall_score(y_test, y_pred,average=None)))
```

```python
from sklearn.ensemble import RandomForestClassifier
param_gridRF = [{
    'n_estimators': [200, 300],
    'max_features': [None, "sqrt"],
    'max_depth': [5, 10, 15]
}]
rf_clf = RandomForestClassifier()
grid_searchRF = GridSearchCV(rf_clf, param_gridRF)
grid_searchRF.fit(X_train, y_train.values.ravel())
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(
    estimators=
        [('lr',LogisticRegression(max_iter=200, multi_class='ovr', random_state=1,solver='newton-cg')),
        ('rf', RandomForestClassifier(max_depth=15, max_features='sqrt', n_estimators=200)),
        ('gbc', GradientBoostingClassifier(max_depth=5, max_features='sqrt', n_estimators=50)),
        ('svm', SVC(kernel='poly', max_iter=800))
        ],voting='hard')
voting_clf.fit(X_train, y_train.values.ravel())
```

```python
y_pred_train = voting_clf.predict(X_train)
print("Accuracy of Ensemble on training: "+str(accuracy_score(y_train, y_pred_train)))
y_pred = voting_clf.predict(X_test)
print("Accuracy of Ensemble on testing: "+str(accuracy_score(y_test, y_pred)))
print("Precision of Ensemble: "+str(precision_score(y_test, y_pred,average=None)))
print("F1 of Ensemble: "+str(f1_score(y_test, y_pred,average=None)))
print("Recall of Ensemble: "+str(recall_score(y_test, y_pred,average=None)))
```
